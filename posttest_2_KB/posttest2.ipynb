#  3.menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

- Membuat Dictionary bernama dataFrameDictionary dengan :
 1. key "numberOfRecords" untuk menyimpan nilai jumlah record
 2. key "numberOfAttributes" untuk menyimpan nilai jumlah attribute
 3. key "AttributeName" untuk menyimpan nilai nama attribute
 4. key "numberOfIntAttribute" untuk menyimpan nilai jumlah attribute dengan tipe angka
- Menampilkan isi value dari dictionary satu per satu (jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka)
- menampilkan isi value dari dictionary sekaligus

```python
dataFrameDictionary = {
  "numberOfRecords": len(dataFrame),
  "numberOfAttributes": len(dataFrame.columns),
  "AttributeName": dataFrame.columns,
  "numberOfIntAttribute" : len(dataFrameInt.columns)
}
# Menampilkan isi value dari dictionary satu per satu
print("jumlah record = ",dataFrameDictionary["numberOfRecords"])
print("\njumlah Attribute = ",dataFrameDictionary["numberOfAttributes"])
print("\nNama Attribute = ",dataFrameDictionary["AttributeName"])
print("\njumlah record dengan tipe Angka = ",dataFrameDictionary["numberOfIntAttribute"])
# menampilkan isi value dari dictionary sekaligus
print("\n------------------------------------------------------------------------------------------------------------------------\n")
print(dataFrameDictionary)
```

# menampilkan jumlah label dari semua attribute yang bernilai object
- Menyimpan dataset(dataFrame) dengan attribute bertipe Object ke dalam variable dataFrameObject
- menggunakan perulangan for untuk menampilkan jumlah label dari Attribute satu per satu

```python
dataFrameObject = dataFrame.select_dtypes('object')
for x in range (0,len(dataFrameObject.columns)):
    print("\nAttribute ->",dataFrameObject.columns[x],"\n")            
    print(dataFrameObject[dataFrameObject.columns[x]].value_counts())
    print("\n===============================================")
```
