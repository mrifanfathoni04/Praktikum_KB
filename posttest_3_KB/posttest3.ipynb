# Visualisasi Data

-mengimport library

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

matplotlib.style.use('ggplot') # grafik menggunakan gaya ggplot
```

import file dataset kedalam variable df

```python
df = pd.read_csv('Clean_Dataset.csv')
df.head()
```

```python
# menghilangkan kolom unnamed:0
df = df.iloc[:,1:]
df.head()
```

# Bar Plot

```python
# pandas
jumlah_penerbangan = df['airline'].value_counts()
jumlah_penerbangan
```

```python
jumlah_penerbangan.plot.bar(title="penerbangan", ylabel="jumlah penerbangan",xlabel='airline')
```

Vistara merupalan perusahaan penerbangan dengan jumlah penerbangan terbanyak dengan lebih dari 100000 penerbangan, Air_india terbanyak kedua, indigo terbanyak ketiga, GO_FIRST terbanyak keempat , AirAsia tebanyak kelima dan spicejet Paling sedikit dengan kurang dari 10000 penerbangan

## Histogram

```python
df['price'].plot.hist(bins=10,range=(0,10000),xlabel="harga", title="penerbangan",ylabel='frekuensi')
```

penerbangan dengan frekuensi distribusi terbanyak merupakan penerbangan dengan range harga 4000-5000 yaitu lebih dari 35000 penerbangan, sedangkan penerbangan dengan frekuensi distribusi paling sedikit merupakan penerbangan dengan range harga 1000-2000 yaitu 5000 penerbangan

## Pie Chart

```python
# pandas
plt.figure(figsize=(15,15))
df['source_city'].value_counts().plot.pie(autopct='%.2f%%',title="kota asal",ylabel='persentase')
```

Delhi merupakan kota dengan jumlah penerbangan terbanyak dengan 20.44%, delhi dan mumbai merupakan dua kota dengan selisih jumlah penerbangan paling sedikit dengan persentase selisih 0.15%

## HEATMAP

```python
int_data=df.select_dtypes('int64').corr()
int_data
```

```python
sns.heatmap(int_data, annot=True).set_title("korelasi data")
```

data price dan days_left berkorelasi negatif (tidak berkorelasi)
